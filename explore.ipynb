```
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```
# Plot number of members vs non-members using the bixi every year

def plot_member(df_2014, df_2015, df_2016, df_2017):
    
    count_2014 = df_2014['is_member'].value_counts()
    count_2015 = df_2015['is_member'].value_counts()
    count_2016 = df_2016['is_member'].value_counts()
    count_2017 = df_2017['is_member'].value_counts()
    
    plt.figure(figsize=(17, 10))
    
    plt.subplot(2,2,1)
    plt.bar(count_2014.index.values, count_2014)
    plt.xticks(count_2014.index.values, ('member', 'non-member'))
    plt.title('Bixi member usage for 2014')
    
    plt.subplot(2,2,2)
    plt.bar(count_2015.index.values, count_2015)
    plt.xticks(count_2015.index.values, ('member', 'non-member'))
    plt.title('Bixi member usage for 2015')
    
    plt.subplot(2,2,3)
    plt.bar(count_2016.index.values, count_2016)
    plt.xticks(count_2016.index.values, ('member', 'non-member'))
    plt.title('Bixi member usage for 2016')
    
    plt.subplot(2,2,4)
    plt.bar(count_2017.index.values, count_2017)
    plt.xticks(count_2017.index.values, ('member', 'non-member'))
    plt.title('Bixi member usage for 2017')
```

```
# Plot duration of bixi usage

def plot_duration(df_2014, df_2015, df_2016, df_2017):
    
    dur_2014 = round(df_2014['duration_sec']/60)
    dur_2015 = round(df_2015['duration_sec']/60)
    dur_2016 = round(df_2016['duration_sec']/60)
    dur_2017 = round(df_2017['duration_sec']/60)
        
    plt.figure(figsize=(17, 10))
    
    plt.subplot(2,2,1)
    plt.hist(dur_2014)
    plt.title('Bixi duration of usage for 2014')
    
    plt.subplot(2,2,2)
    plt.hist(dur_2015)
    plt.title('Bixi duration of usage for 2015')
    
    plt.subplot(2,2,3)
    plt.hist(dur_2016)
    plt.title('Bixi duration of usage for 2016')
    
    plt.subplot(2,2,4)
    plt.hist(dur_2017)
    plt.title('Bixi duration of usage for 2017')
```

```
# Plot temperature-based bixi usage

def plot_temperature(df_2014, df_2015, df_2016, df_2017):
        
    plt.figure(figsize=(17, 10))
    
    plt.subplot(2,2,1)
    plt.hist(df_2014['temperature'])
    plt.title('Bixi temperature-based usage for 2014')
    
    plt.subplot(2,2,2)
    plt.hist(df_2015['temperature'])
    plt.title('Bixi temperature-based usage for 2015')
    
    plt.subplot(2,2,3)
    plt.hist(df_2016['temperature'])
    plt.title('Bixi temperature-based usage for 2016')
    
    plt.subplot(2,2,4)
    plt.hist(df_2017['temperature'])
    plt.title('Bixi temperature-based usage for 2017')
```

```
# Plot usage duration based on temperature

def plot_dur_temp(df_2014, df_2015, df_2016, df_2017):
    
    count_2014 = df_2014.groupby(['temperature']).count()['duration_sec']
    count_2015 = df_2015.groupby(['temperature']).count()['duration_sec']
    count_2016 = df_2016.groupby(['temperature']).count()['duration_sec']
    count_2017 = df_2017.groupby(['temperature']).count()['duration_sec']
    
    plt.figure(figsize=(17, 10))
    
    plt.subplot(2,2,1)
    plt.bar(count_2014.index.values, count_2014)
    plt.title('Bixi temperature-based usage duration for 2014')
    
    plt.subplot(2,2,2)
    plt.bar(count_2015.index.values, count_2015)
    plt.title('Bixi temperature-based usage duration for 2015')
    
    plt.subplot(2,2,3)
    plt.bar(count_2016.index.values, count_2016)
    plt.title('Bixi temperature-based usage duration for 2016')
    
    plt.subplot(2,2,4)
    plt.bar(count_2017.index.values, count_2017)
    plt.title('Bixi temperature-based usage duration for 2017') 
```
